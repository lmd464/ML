```python
import numpy as np
from matplotlib import pyplot as plt


train_data_input = np.loadtxt("C:\\Users\\l4m6d4\\Desktop\\3-2\\기계학습\\ML_03_201702081_최재범\\iris_train.csv", 
                        delimiter = ',', 
                        dtype = np.float32)
test_data_input = np.loadtxt("C:\\Users\\l4m6d4\\Desktop\\3-2\\기계학습\\ML_03_201702081_최재범\\iris_test.csv", 
                        delimiter = ',', 
                        dtype = np.float32)[:, 0:4]


train_data_1 = train_data_input[0:40, 0:4]
train_data_2 = train_data_input[40:80, 0:4]
train_data_3 = train_data_input[80:120, 0:4]





# 1. 평균 벡터 계산

# class 1
column_sum_1 = np.sum(train_data_1, axis = 0)
avg_vector_1 = column_sum_1 / train_data_1.shape[0]

# class 2
column_sum_2 = np.sum(train_data_2, axis = 0)
avg_vector_2 = column_sum_2 / train_data_2.shape[0]

# class 3
column_sum_3 = np.sum(train_data_3, axis = 0)
avg_vector_3 = column_sum_3 / train_data_3.shape[0]


print("1번 군집의 평균벡터")
print(avg_vector_1)
print()
print("2번 군집의 평균벡터")
print(avg_vector_2)
print()
print("3번 군집의 평균벡터")
print(avg_vector_3)
print()





# 2. 평균 벡터와 테스트 데이터의 거리 계산
def dist(avg_vector, test_data):
    
    data_amount = test_data.shape[0]
    dist_list = np.zeros(data_amount)
    
    for data_num in range(0, data_amount):
        sub = test_data[data_num] - avg_vector
        sum = np.sum(sub ** 2)
        dist = sum ** 0.5
        dist_list[data_num] = dist
    
    return dist_list


dist_1_list = dist(avg_vector_1, test_data_input)
dist_2_list = dist(avg_vector_2, test_data_input)
dist_3_list = dist(avg_vector_3, test_data_input)
print("테스트 데이터의 1번 군집과의 거리")
print(dist_1_list)
print()
print("테스트 데이터의 2번 군집과의 거리")
print(dist_2_list)
print()
print("테스트 데이터의 3번 군집과의 거리")
print(dist_3_list)
print()





# 3. 거리값에 따라 분류
def classify(dist_1_list, dist_2_list, dist_3_list):
    
    classify_result_list = []
    data_amount = test_data_input.shape[0]
    
    for data_num in range(0, data_amount):
        
        # 1번 군집에 가까움
        if dist_1_list[data_num] < dist_2_list[data_num] and dist_1_list[data_num] < dist_3_list[data_num]:
            classify_result_list.append(1)
        
        # 2번 군집에 가까움
        elif dist_2_list[data_num] < dist_1_list[data_num] and dist_2_list[data_num] < dist_3_list[data_num]:
            classify_result_list.append(2)
        
        # 3번 군집에 가까움
        elif dist_3_list[data_num] < dist_1_list[data_num] and dist_3_list[data_num] < dist_2_list[data_num]:
            classify_result_list.append(3)
        
        else:
            classify_result_list.append(-999)
    
    return classify_result_list
    

classify_result_list = classify(dist_1_list, dist_2_list, dist_3_list)
print("분류 결과")
print(classify_result_list)
print()
    
    
    
    
    
# 4. 실제 군집값과의 비교 후, 정확도 반환
def accurancy(classify_result_list):
    
    data_amount = len(classify_result_list)
    real_classification = np.zeros(data_amount)
    real_classification[0:10] = 1
    real_classification[10:20] = 2
    real_classification[20:30] = 3
    
    correct_count = 0
    for classify_result_count in range(0, data_amount):
        if classify_result_list[classify_result_count] == real_classification[classify_result_count]:
            correct_count = correct_count + 1
               
    return correct_count / data_amount 


accurancy = accurancy(classify_result_list)
print("정확도 : %f" % accurancy)
    
```
