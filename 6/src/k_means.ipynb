```python
from sklearn.datasets import make_moons

from sklearn.cluster import KMeans
from matplotlib import pyplot as plt

X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=60, edgecolors='k')



# N_CLUSTER
kmeans = KMeans(n_clusters=10)
kmeans.fit(X)
y_pred = kmeans.predict(X)


plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
           marker='^', s=100, linewidth=2, edgecolors='k')
plt.title("clusters = 10")
```

```python
from sklearn.datasets import make_blobs

# CLUSTER_STD
X_varied, y_varied = make_blobs(n_samples = 200, cluster_std = 0.5, random_state=170)
plt.scatter(X_varied[:, 0], X_varied[:, 1], s=60, edgecolors='k')



# N_CLUSTER
kmeans_2 = KMeans(n_clusters = 10, random_state = 0)
kmeans_2.fit(X_varied)
y_pred = kmeans_2.predict(X_varied)


plt.scatter(kmeans_2.cluster_centers_[:, 0], kmeans_2.cluster_centers_[:, 1], 
           marker='^', s=100, linewidth=2, edgecolors='k')
plt.title("clusters = 7, cluster_std = 0.5")
```
